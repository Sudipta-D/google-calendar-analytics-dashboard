```python
query = """
SELECT 
    event_type,
    COUNT(*) AS total_events,
    SUM(
        (julianday(end_time) - julianday(start_time)) * 24 * 60
    ) AS total_minutes
FROM calendar_events
GROUP BY event_type
ORDER BY total_minutes DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    strftime('%Y-%W', start_time) AS week,
    COUNT(*) AS events
FROM calendar_events
GROUP BY week
ORDER BY week;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    ROUND(
        SUM(CASE WHEN event_type = 'Project Work' THEN 
            (julianday(end_time) - julianday(start_time)) * 24 * 60 ELSE 0 END) * 100.0 /
        SUM((julianday(end_time) - julianday(start_time)) * 24 * 60), 2
    ) AS focus_percent
FROM calendar_events;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    DATE(start_time) AS date,
    ROUND(SUM((julianday(end_time) - julianday(start_time)) * 24), 2) AS total_hours
FROM calendar_events
GROUP BY DATE(start_time)
HAVING total_hours > 8
ORDER BY total_hours DESC;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    DATE(start_time) AS date,
    ROUND(SUM((julianday(end_time) - julianday(start_time)) * 24), 2) AS total_hours
FROM calendar_events
GROUP BY DATE(start_time)
HAVING total_hours > 0
ORDER BY total_hours DESC;
"""

pd.read_sql(query, conn)
```
